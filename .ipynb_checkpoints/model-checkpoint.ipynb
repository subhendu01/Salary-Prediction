```python
# import the library
import pandas as pd
import pickle
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.rename(columns = {'test_score(out of 10)': 'test_score', 'interview_score(out of 10)': 'interview_score'}, inplace = True)
df
```

```python
df['experience'].fillna(0,inplace=True)
df
```

```python
# converting experience  into integer
def con_to_int(value):
    word_dict = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8,
                'nine':9, 'ten':10, 'eleven':11, 'twelve':12, 'zero':0, 0: 0}
    return word_dict[value]
```

```python
df['experience'] = df['experience'].apply(lambda x: con_to_int(x))
df
```

```python
# replacing na test_score to mean value
df['test_score'].fillna(df['test_score'].mean(), inplace = True)
df
```

```python
# Independent Variable
X = df.iloc[:,:-1]
X
```

```python
# Dependent Variable
y = df.iloc[:,-1:]
y
```

```python
#Splitting Training and Test Set
#Since we have a very small dataset, we will train our model with all availabe data.
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
#Fitting model with trainig data
regressor.fit(X, y)
```

```python
# Saving model to disk
pickle.dump(regressor, open('model.pkl','wb'))
```

```python
# Loading model to compare the results
model = pickle.load(open('model.pkl','rb'))
print(model.predict([[2, 9, 6]]))
```
